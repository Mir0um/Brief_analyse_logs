```python
with open("warning.txt", 'r') as f:
    dangerous_ips = set(line.strip() for line in f)
```

```python
with open('connexion.log', 'r') as fco:
    with open('user.txt', 'w') as fus:
        for line in fco:
            fus.write(f"{line.strip().split(';')[1]};\n")
            print(line.strip().split(';')[1])
```

```python
suspect_users = {}
with open('connexion.log', 'r') as f:
    for line in f:
        ip, user, _ = line.strip().split(';')
        if ip in dangerous_ips:
            suspect_users[user] = suspect_users.get(user, 0) + 1
```

```python
print(f"Voici la liste des suspects: {suspect_users}")
```

```python
with open('suspect.txt', 'w') as f:
    for user, count in sorted(suspect_users.items(), key=lambda x: x[1], reverse=True):
        f.write(f"{user};{count}\n")
```
